### Tabulation
#### One-Way Tables
Create frequency tables (also known as crosstabs) in pandas using the _pd.crosstab()_ function.The function takes one or more array-like objects as indexes or columns and then constructs a new DataFrame of variable counts based on the supplied arrays.These simple one-way tables give us some useful insight: we immediately get a sense of distribution of records across the categories.

__Excercise:__ 
 - Show the number hospitals in each district and subdistrict.
         Hint:Do a one way table for distirct,sub_district and sector --> can we plot this??    
 - A cross tab of monitoring_evaluation:S0_21 


```python
## this creates a summary table of hosptoals in each province
province_tab = pd.crosstab(index=iss_df["facility_identification:province"],  # Make a crosstab
                              columns="count")      # Name the count column

province_tab
```

One of the most useful aspects of frequency tables is that they allow you to plot and extract the proportion of the data that belongs to each category . With a one-way table, to get the proportion you can do this by dividing each table value by the total number of records in the table:

#### Two-Way Tables
Two-way frequency tables, also called contingency tables, are tables of counts with two dimensions where each dimension is a different variable. Two-way tables can give you insight into the relationship between two variables. To create a two way table, pass two variables to the _pd.crosstab()_ function instead of one:

__Excercise:__ 
 - Show the number and proportion  of facilities in each district, subdistrict and sector that are urban or rural
     Hint: Do a two way table of facility_identification:S0_13 and district variable --> can we plot this??    
 - Show the number and proportion  of facilities in each district and subdistrict by their type (facility_identification:S0_12) 

```python
# Summary of hospitals in a province that are urban or rural
# Table of province vs. S0_13
prov_S0_13 = pd.crosstab(index=iss_df["facility_identification:province"], 
                           columns=iss_df["facility_identification:S0_13"])

prov_S0_13
```

The margins keyword instructed pandas to add a total for each row as well as a total at the bottom. I also passed a value to margins_name in the function call because I wanted to label the results “Total” instead of the default “All”.

```python
pd.crosstab(iss_df["facility_identification:province"], iss_df["facility_identification:S0_13"],
            margins=True, margins_name="Total")
```

```python
## stacked bar plot of the counts
prov_S0_13.plot.bar(stacked=True)
```

```python
## stacked bar plot of the counts
prov_S0_13.plot.bar(stacked=False, rot=0)
```

crosstab allows us to do even more summarization by including values to aggregate. To illustrate this, we can calculate the average __"monitoring_evaluation4:S0_86"__ of facilities by province and facility type:
By using _aggfunc='mean'_ and _"values=iss_df["monitoring_evaluation4:S0_86"]"__ we are telling pandas to apply the mean function to the__"monitoring_evaluation4:S0_86"__ of all the combinations of the cross tab. In those provinces where there is no __monitoring_evaluation4:S0_86 recorded__, it displays NaN

```python
pd.crosstab(iss_df["facility_identification:province"], iss_df["facility_identification:S0_13"], 
            values=iss_df["monitoring_evaluation4:S0_86"], aggfunc='mean').round(2)
```

To get the total proportion of counts in each cell, divide the table by the grand total:

```python
prov_row_crosstab = prov_S0_13.apply(lambda r: r/r.sum(), axis=1)*100
prov_row_crosstab
```

```python
## stacked bar plot of the counts
prov_row_crosstab.plot.bar(stacked=True)
```

```python
## quick way of creating percentage using normalize
prov_S0_13_prop = pd.crosstab(index=iss_df["facility_identification:province"], 
                           columns=iss_df["facility_identification:S0_13"],
                        normalize="index")

prov_S0_13_prop
```

```python
#proportion by column
prov_col_crosstab = prov_S0_13.apply(lambda r: r/r.sum(), axis=0)*100
prov_col_crosstab
```
